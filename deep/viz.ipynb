```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
best_mult = np.load('best_mult.npy')
best_p_val = np.load('best_pval.npy')
```

```python
print(best_mult, best_p_val)
```

```python
all_jump_mults = np.load('all_mults.npy')
all_mean_p_values = np.load('all_means.npy')
```

```python
plt.figure(figsize=(8, 5))
for i in range(len(all_jump_mults)):
    plt.errorbar(all_jump_mults[i], all_mean_p_values[i], 
                fmt='o-', capsize=5, label=f'Iteration {i+1}')
plt.xlabel('Jump Multiplier', fontsize=15, labelpad=15)
plt.ylabel('Mean $p$-value', fontsize=15, labelpad=10)
plt.tick_params(axis='x', labelsize=12)
plt.tick_params(axis='y', labelsize=12)
plt.grid(True)
plt.legend()
plt.savefig('grid_search_shallow_2_deep.svg', transparent=True)
plt.show()
```
